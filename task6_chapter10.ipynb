# 面向列的多函数应用

```python
tips = pd.read_csv('examples/tips.csv')
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips[:6]
```

```python
grouped = tips.groupby(['day','smoker']) #根据什么分组会输出分组名
grouped_pct = grouped['tip_pct']
grouped_pct.agg('mean')  #聚合函数
```

```python
grouped_pct.agg(['mean','std',peak_to_peak])
```

```python
grouped_pct.agg([('foo','mean'),('bar',np.std)])  #传入的是name和function组成的列表
```

```python
functions = ['count', 'mean', 'max']
result = grouped['tip_pct','total_bill'].agg(functions) #对grouped的两个列操作
result
```

```python
result['tip_pct']
```

```python
ftuple = [('Durchschnitt', 'mean'), ('Abweichung', np.var)]
grouped['tip_pct','total_bill'].agg(ftuple) #变量名后的列表为输出值，后面为聚合函数
```

```python
grouped.agg({'tip':np.max,'size':'sum'})  #输出所有列的值
```

```python
grouped.agg({'tip_pct':['min','max','mean','std'],
             'size':'sum'})
```

# 分位数和桶分析

```python
frame = pd.DataFrame({'data1':np.random.randn(1000),
                      'data2':np.random.randn(1000)})
```

```python
quartiles = pd.cut(frame.data1,4) #指定一列分桶
quartiles[:10]
```

```python
def get_stats(group):
    return {'min': group.min(), 'max': group.max(),
            'count': group.count(), 'mean': group.mean()}
grouped = frame.data2.groupby(quartiles)
grouped.apply(get_stats).unstack()  #学习！！！！
```

```python
grouped.apply(get_stats)
```

```python
grouping  = pd.qcut(frame.data1,10,labels=False)
grouped = frame.data1.groupby(grouping)
grouped.apply(get_stats).unstack()
```

# 随机采样和排列

```python
suits = ['H', 'S', 'C', 'D']
card_val = (list(range(1, 11)) + [10] * 3) * 4 #“+”代表连接
base_names = ['A'] + list(range(2, 11)) + ['J', 'K', 'Q']
cards = []
for suit in ['H', 'S', 'C', 'D']:
    cards.extend(str(num) + suit for num in base_names)  #str(num)的取值？？内循环str(num)是取字符串不是长度

deck = pd.Series(card_val, index=cards)
```

```python
deck[:13]
```

```python
def draw(deck,n=5):
    return deck.sample(n)
```

```python
draw(deck)
```

```python
get_suit = lambda card :card[-1]
deck.groupby(get_suit).apply(draw,n=2)
```

```python
deck.groupby(get_suit,group_keys = False).apply(draw,n=2)
```

# 元组级别的线性回归

```python
import statsmodels.api as sm
def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept'] = 1.  #???
    result = sm.OLS(Y, X).fit()
    return result.params
by_year.apply(regress,'AAPL',['SPX'])
```
